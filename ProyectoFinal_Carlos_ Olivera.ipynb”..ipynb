```python
# Contar la cantidad de IsActiveMember por g√©nero
active_members = df_Churn_Modelling_archivo[df_Churn_Modelling_archivo['IsActiveMember'] == 1]
inactive_members = df_Churn_Modelling_archivo[df_Churn_Modelling_archivo['IsActiveMember'] == 0]

active_gender_counts = active_members['Gender'].value_counts()
inactive_gender_counts = inactive_members['Gender'].value_counts()

# Calcular la diferencia entre activos e inactivos para cada g√©nero
gender_differences = {}
for gender in ['Male', 'Female']:
    active_count = active_gender_counts.get(gender, 0)
    inactive_count = inactive_gender_counts.get(gender, 0)
    difference = active_count - inactive_count
    gender_differences[gender] = difference

# Crear el gr√°fico de barras
fig, ax = plt.subplots(figsize=(10, 6))
x = np.arange(len(active_gender_counts.index))
width = 0.4

active_bars = ax.bar(x - width/2, active_gender_counts.values, width, label='Activos')
inactive_bars = ax.bar(x + width/2, inactive_gender_counts.values, width, label='Inactivos')

# Agregar texto con las proporciones dentro de las barras
for i, (active_count, inactive_count) in enumerate(zip(active_gender_counts.values, inactive_gender_counts.values)):
    total_count = active_count + inactive_count
    active_proportion = active_count / total_count
    inactive_proportion = inactive_count / total_count
    
    ax.text(x[i] - width/2, active_count/2, f'{active_proportion:.2%}', ha='center', va='center', color='white', fontsize=10)
    ax.text(x[i] + width/2, inactive_count/2, f'{inactive_proportion:.2%}', ha='center', va='center', color='white', fontsize=10)

# Agregar leyenda con las diferencias
legend_labels = []
for gender, difference in gender_differences.items():
    if difference > 0:
        label = f"{gender}: +{difference}"
    else:
        label = f"{gender}: {difference}"
    legend_labels.append(label)

ax.set_xticks(x)
ax.set_xticklabels(active_gender_counts.index)
ax.set_xlabel('G√©nero')
ax.set_ylabel('N√∫mero de Personas')
ax.set_title('G√©nero de Clientes Activos vs Inactivos')
ax.legend(labels=legend_labels, loc='center left', bbox_to_anchor=(1, 0.5))

plt.show()
```
